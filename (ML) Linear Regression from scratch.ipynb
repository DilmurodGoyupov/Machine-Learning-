 # <center> Linear Regression - The goal of linear regression is to find the best line that fits the given points

 * The blue line should be adjusted in such a way that the distance from it to the elements on both sides (functional poter, loss) should be as small as possible.




The formula for linear regression is as follows

# <center> <h1>  `y = ax + b`



`a = deflection coefficient (weight)`

`b = point of intersection of the line with the y-axis (bias)`

`x = ourpoints(data_x)`

We can find a and b by least mean square method and Gradient descent

# <center> Linear Regression.

## <center> 1 - Least Mean Square Method

find the coefficient a.

a = sum(from the current point x_train - the average of all x_trains is subtracted) * (from the current y_train - the average of all y_trains is subtracted) / sum(from the point x_train - the average of all x_trains is subtracted)**2




Find the point of intersection (bias) of the b line with the y axis

b = (average of all y_trains - a * y_train)



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

`We use a smaller dataset. Our dataset contains cars name, model, height, weight and CO2. In Linear we only use [Weight CO2] In Multi Linear we use [Volume Weight CO2].`

```python
data = pd.read_csv("Datasets/datat.csv")
data
```

```python
# Since we're only doing linear regression now, we'll only get the Weight and CO2 column
x = data[["Weight"]].to_numpy()
y = data[["CO2"]].to_numpy()
```

```python
x
```

```python
y
```

```python
def covariance(x,y):
    mx = x - x.mean()
    my = y - y.mean()
    
    covar = sum(mx * my)
    var = sum(mx**2)
    
    return covar / var
```

```python
def calculate(x,y):
    a = covariance(x,y)
    b = y.mean() - a * x.mean()
    return a,b
```

```python
weight, bias = calculate(x,y)
weight, bias
```

```python
# Now let's try to predict. So if the weight of the car is 790
weight * x[0] + bias
# 93.4 is predicted to come out
```

#### We can know how good our prediction is by MSE(Mean squared error) or MAE(Mean absolute error)





```python
# predict
y_bash = weight * x + bias
y_bash
```

```python
mse = (y - y_bash)**2
mae = abs(y - y_bash)
```

```python
sum(mse) / len(data)
# MSE
```

```python
sum(mae) / len(data)
# MAE
```

` The summary assessment is not good but our date and feauture are lacking. This is natural.`

##  <center> 2 - Multi Linear Regression

Multi Linear Regression is slightly different from linear. since we have a lot of x's, we work by matrix multiplication.



```python
data = pd.read_csv("Datasets/datat.csv")
data
```

```python
# we take Volume, Weight columns as x and CO2 column as y
x = data[["Volume", "Weight"]].to_numpy()
y = data[["CO2"]].to_numpy()
```

```python
x
```

```python
y
```

```python
#We find the average of column 1 of our x value.
x1_mean = x[:,0].mean()
x1_mean
```

```python
# We find the average of column 2 of our x value.
x2_mean = x[:,1].mean()
x2_mean
```

```python
# We find the mean of the column of y
y_mean = y.mean()
y_mean
```

```python
x[:,0] = x[:,0] - x1_mean
# first, its average value is subtracted from the incoming x1 value
```

```python
x[:,1] = x[:,1] - x2_mean
# second, its average value is subtracted from the incoming x2 value
```

```python
y_var = y - y_mean
y_var
# third, its average value is subtracted from the y value
```

```python
x
```

```python
y_var
```



We multiply the x matrix that is the transpose by the x matrix. Then we get the inverse matrix.

```python
inv_multiplication = np.linalg.inv(np.dot(x.T,x))
inv_multiplication
```



We multiply the value found above by Transpose x and y_var.

```python
weights = inv_multiplication.dot(x.T).dot(y_var)
weights
```

```python
bias = y_mean - weights[0] * x1_mean - weights[1] * x2_mean
bias
```

```python
predict = weights[0] * 100 + weights[1]*790 + bias
predict
```

```python
y_hat = []
for i in range(len(x)):
    y_hat.append(sum(weights[0] * x[i][0] + weights[1]*x[i][1] + bias))
```

```python
y_hat = np.array(y_hat)
y_hat
```

```python
y_hat = np.expand_dims(y_hat,1)
y_hat
```

```python
y
```

```python
mse = (y - y_hat)**2
mae = abs(y - y_hat)
```

```python
sum(mse) / len(data)
```

```python
sum(mae) / len(data)
```

` The bottom line is not good. We'll see with SGD `

## <center> 3- Linear Regression (SGD)

```python
data = pd.read_csv("Datasets/datat.csv")
data
```

```python
x = data[["Weight"]].to_numpy()
y = data[["CO2"]].to_numpy()
```

```python
# we can normalize the data, because very large values may appear during the calculation process.
x = x/x.max()
x
```

```python
# We also normalize y.
y = y / y.max()
y
```

```python
lr = 0.001      # learning rate
epochs = 10000  # epochs
w = 0.8         # initial value of weight (we got random)
b = 0.5         # initial value of bias (we got random)
```

```python
for epoch in range(epochs):
    for x_i, y_i in zip(x,y):
        
        y_bash = w*x_i + b
        
        loss = (y_i - y_bash)**2
        
        w = w - 0.001 * (2 * x_i * (y_bash - y_i))
        b = b - 0.001 * (2 * (y_bash - y_i))
        
        
    if epoch % 100 ==0:
        print(f"Epoch - {epoch}: Loss : {loss}")
        print("Weight = ",w, "  Bias = ", b)
        
```

```python
#prdict
lr = w*0.45246277+b
lr
```

```python
loss
#mse
```

#### Linear regression is not bad via SGD

## <center> 4 - Multi Linear Regression (SGD)

```python
data = pd.read_csv("Datasets/datat.csv")
data
```

```python
x = data[["Volume", "Weight"]].to_numpy()
y = data[["CO2"]].to_numpy()
```

```python
x
```

```python
x1=[]
x2=[]

for i in range(len(x)):
    
    x1.append(x[i][0])
    x2.append(x[i][1])
```

```python
x1 = x1/max(x1)
x1
# data normalization
```

```python
x2 = x1/max(x2)
x2
# data normalization
```

```python
y = y / max(y)
y
# data normalization
```

```python
w1=-2          # initial value of weight1 (we got random)
w2=3           # initial value of weight2 (we got random)

b1=1.5         # initial value of bias (we got random)

epochs=10000   # number of epochs

lr=0.001       # learning rate
```

```python
for epoch in range(epochs):
    for x1_i, x2_i, y_i in zip(x1,x2,y):
        
        y_bash = w1*x1_i + w2*x2_i + b1
        
        loss = (y_i - y_bash)**2
        
        w1 = w1 - 0.001 * (2 * x1_i * (y_bash - y_i))
        w2 = w2 - 0.001 * (2 * x2_i * (y_bash - y_i))
        
        b1 = b1 - 0.001 * (2 * (y_bash - y_i))
       
        
        
    if epoch % 100 ==0:
        print(f"Epoch - {epoch}: Loss : {loss}")
        print("Weight = ",w1,w2, "  Bias = ", b1)
        
```

```python
w1*0.45 + w2*0.0002291 + b1
```

```python
loss
#mse
```
