`First we see the probability that an unknown fruit is a banana.`

### `Let's check if our unknown fruit is a banana.`

* 1. Unknown fruit length = short, so we calculate the value in our table 1
---------------

* 2. Unknown fruit sweetness = sweet, so we calculate the value in our table 2
---------------
* 3. Unknown fruit yellowness = yellow means we calculate the value in our table 3
---------------
* 4 - We calculate the probability for the banana class.



```python
# 1 - P(Length | Banana)
L_B = 100 / 500
L_B

# Number of bananas that are not 100 long
# 500 total number of bananas
```

```python
# 2 - P(Sweet | Banana)
S_B = 350 / 500
S_B

# 350 sweet bananas count
# 500 total number of bananas
```

```python
# 3 - P(Yellow | Banana)
Y_B = 450 / 500
Y_B

# 450 number of yellow bananas
# 500 total number of bananas
```

```python
# 4 - P(Banana)
P_B = 500 / 1000
P_B

# 500 total number of bananas
# 1000 total number of fruits
```

```python
# P(Banana | Length, Sweet, Yellow)
probability_of_being_banana = L_B * S_B * Y_B * P_B
probability_of_being_banana
```

` We've only considered the possibility of a banana, now we need to consider the possibility of an Orange.`



### `Let's check if our unknown fruit is orange.`

* 1. Unknown fruit length = short, so we calculate the value in our table 1

---------------

* 2. Unknown fruit sweetness = sweet, so we calculate the value in our table 2

---------------

* 3. Unknown fruit yellowness = yellow means we calculate the value in our table 3

---------------

* 4. We calculate the probability of the orange class




```python
# 1 - P(Length | Orange)
L_A = 300 / 300
L_A

# 300 is not the number of long oranges
# 300 total number of oranges
```

```python
# 2 - P(Sweet | Orange)
S_A = 150 / 300
S_A

# 150 number of sweet oranges
# 300 total number of oranges
```

```python
# 3 - P(Yellow | Orange)
Y_A = 300 / 300
Y_A

# 300 is the number of yellow oranges
# 300 total number of oranges
```

```python
# 4 - P(Orange)
P_A = 300 / 1000
P_A

# 300 total number of oranges
# 1000 total number of fruits
```

```python
# P(Banana | Length, Sweet, Yellow)
probability_of_being_orange = L_A * S_A * Y_A * P_A
probability_of_being_orange
```

`We have considered the possibility of Bananas and Apples, now we need to consider the possibility of another fruit.`



### `Let's check if our unknown fruit is another fruit. `

```python
# 2 - P(Sweet | Others)
S_A = 150 / 200
S_A

# 150 number of sweet other fruits
# 200 total number of other fruits
```

```python
# 3 - P(Yellow | Others)
Y_A = 50 / 200
Y_A

# 50 is the number of yellow other fruits
# 200 total number of other fruits
```

```python
# 4 - P(Anothers)
P_A = 200 / 1000
P_A

# 200 total number of other fruits
# 1000 total number of fruits
```

```python
# P(Others | Length, Sweet, Yellow)
others_probability = L_A * S_A * Y_A * P_A
others_probability
```

```python
print("Banana: ", probability_of_being_banana, ".  Orange: ", probability_of_being_orange, ".  Another fruits", others_probability)
print("Probabilities of being")
```
