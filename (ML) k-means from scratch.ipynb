### <center>k-means clustering is a method of vector quantization, originally from signal processing, that aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean (cluster centers or cluster centroid), serving as a prototype of the cluster. 

### <center> k-means allows us to divide our dataset into several clusters (groups). That is, similar data are in one cluster. First, we randomly select the centroid (not too random, as you can see in the code). Then we calculate the distance from each data to the centroid (for example, Euclidean).



### The centroid that the data is closest to belongs to that centroid. Then, in each cycle, we change the centroid by comparing it with the previous value and calculate the distance from the data to the centroid again.



```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# Create random data points
a1 = np.random.randint(1,10, size=(4,2))
a2 = np.random.randint(7,15, size=(4,2))
a3 = np.random.randint(12,20, size=(4,2))
```

```python
x = np.concatenate((a1,a2,a3),axis=0)
plt.plot(x[:,0],x[:,1],".")
# show data
```

```python
x
```

```python
M = np.mean(x, axis=0)      # we find the averages on each axis (x,y).
M
```

```python
D = np.var(x, axis=0)       # we calculate the variance on each axis (x,y).
D
```

```python
K = 3
```

```python
ma = [np.random.normal(M, np.sqrt(D / 10), 2) for n in range(K)]        # initial centers of clusters
ma
# let's take k random centroids
```

```python
COLORS = ('green', 'blue', 'brown', 'black')
```

```python
X = [[] for i in range(K)]
X
```

```python
def euclid(x,y):
    
    evklid = np.mean((x - y) ** 2)
    
    return evklid
```

### 1-cycle

```python
for i in x:
    r = [euclid(i,m) for m in ma] # we calculate the distance from each point to the centroid
    X[np.argmin(r)].append(i)     # returns the index of the minimum value
```

```python
X
```

```python
ma = [np.mean(xx, axis=0) for xx in X]
# update centroids 
```

```python
ma
```

```python
# display of found clusters
for i in range(K):
    xx = np.array(X[i]).T
    plt.scatter(xx[0], xx[1], s=10, color=COLORS[i])

# display of cluster centers
mx = [m[0] for m in ma]
my = [m[1] for m in ma]
plt.scatter(mx, my, s=50, color='red')

plt.show() 
```

### `When should we stop the cycle? `
### `we can stop the loop if our sentoids stop changing or if the change is too low`
### `we'll do this below`

```python
# Create random data points
a1 = np.random.randint(1,10, size=(10,2))
a2 = np.random.randint(7,15, size=(10,2))
a3 = np.random.randint(12,20, size=(10,2))
```

```python
a3
```

```python
x = np.concatenate((a1,a2,a3),axis=0)
plt.plot(x[:,0],x[:,1],".")
# show data
```

```python
x
```

```python
M = np.mean(x, axis=0)      # we find the averages on each axis (x,y).
M
```

```python
D = np.var(x, axis=0)       # we calculate the variance on each axis (x,y).
D
```

```python
K = 3
```

```python
ma = [np.random.normal(M, np.sqrt(D / 10), 2) for n in range(K)]        # initial centers of clusters
ma
# let's take k random centroids
```

```python
def euclid(x,y):
    
    evklid = np.mean((x - y) ** 2)
    
    return evklid
```

```python
COLORS = ('green', 'blue', 'brown', 'black')
```

```python
x_true = True

n=0



while x_true==True:
    
    X = [[] for i in range(K)] 
    
    for i in x:
        r = [euclid(i, m) for m in ma]         # we calculate the distance from each point to the centroid
        X[np.argmin(r)].append(i)          # returns the index of the minimum value
        
        
    ma = [np.mean(xx, axis=0) for xx in X]      # update centroids
    
    
     # We compare the position of the previous centroid and the new one and stop the cycle if the change is less than 0.001
     # or stop the cycle when the cycle reaches 500
    
    if n==0:
        ma_cp=np.zeros((K,2))
        ayirma_cp = np.zeros(K)
        ayirma=[]
        for i in range(K):
            ayirma1 = ((ma[i][0] - ma_cp[i][0])**2 + (ma[i][1] - ma_cp[i][1])**2)
            ayirma.append(ayirma1)
            
        
        ma_cp=ma
        ayirma_cp = ayirma
    
    else:
        ayirma=[]
        for i in range(K):
            ayirma1 = ((ma[i][0] - ma_cp[i][0])**2 + (ma[i][1] - ma_cp[i][1])**2)
            
            ayirma.append(ayirma1)
        s=0
        
        for i in range(K):
            
            
            if abs(ayirma[i]-ayirma_cp[i])<0.00001:   # Threshold
                s+=1
        
        if s==3:
            x_true=False
            
        
        ma_cp=ma
        ayirma_cp=ayirma
    
    n+=1
    
    if n==500:
        x_true=False
        print("We stopped because the cycle reached 500")
    
        
        
```

```python
# display of found clusters
for i in range(K):
    xx = np.array(X[i]).T
    plt.scatter(xx[0], xx[1], s=10, color=COLORS[i])

# display of cluster centers
mx = [m[0] for m in ma]
my = [m[1] for m in ma]
plt.scatter(mx, my, s=50, color='red')

plt.show()  
```
